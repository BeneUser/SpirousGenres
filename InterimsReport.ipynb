#### Possible Audio Data Augmentation

Based on: [Audio Data Augmentation Techniques: The Theory by Valerio Velardo - The Sound of AI](https://www.youtube.com/watch?v=bm1cQfb_pLA)


Data augmentation in audio data can be applied in two ways: We can either augment the waveform or do transforamtions in the spectogram (time-frequency) representation. 

Valerio Velerdo presents a vast amount of possible approaches of which not all apply well to music. His top two methods for music are time stretching (change the speed of the sound without changing the pitch) and pitch scaling (changing the pitch without changing the frequency). Other methods he mentions that work reasonably well for music are a noise addition (white, pink or background noise), impulse response addition (adding different types of reverbs). 
With all methods of data augmentation, we cannot distort the audio too much because if we do so, the audio snippet might not be representitive for its genre anymore. 


At this point it needs to be seen what kind of data augmentiontion we want to try to (hopefully) improve the classifiction of the genres.
